#### Going to get books title starts with 's' from 10 pages

```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Optional: set Firefox to run in headless mode
options = Options()
# options.add_argument("--headless")  # Uncomment to run without opening the browser window

# Set up the WebDriver (assumes geckodriver is in your PATH)
service = Service()
driver = webdriver.Firefox(service=service, options=options)

driver.get("http://127.0.0.1:5500/project/index.html")
print(driver.title)
# search = driver.find_element(By.NAME,'s')
search = driver.find_element(By.ID,'search-input')
search.clear() # Clear any existing text in the search input
search.send_keys("ras")
# search.send_keys(Keys.RETURN)
print(driver.page_source) # Entire page source after search input
time.sleep(6)
driver.quit()
```

```python

# wait until the presence of a specific element 
try:
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "search-results"))
    )
    print("Element found:", element.text) # element.text will give the text content of the element
except:
    driver.quit()

articles = element.find_elements(By.TAG_NAME, "article")
for article in articles:
    title = article.find_element(By.TAG_NAME, "h2").text
    description = article.find_element(By.TAG_NAME, "p").text
    print(f"Title: {title}\nDescription: {description}\n")
```
